```python
df = df.join(pd.DataFrame(dfF))
df.head()
```

```python
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense

drop=['id', 'toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate', 'NumC', 'NumL', 'Positive']

X=df.drop(drop, axis=1)
print('Features---')
print(X.head())
y=df[target]
print(target)
print(y.head())
y_categorical = to_categorical(y)
```

```python
input_dim = X.shape[1]
toxic_model = Sequential()
#toxic_model.add(Dense(units=input_dim * 2, activation='relu', input_dim=input_dim))

toxic_model.add(Dense(200, activation='relu', input_dim=input_dim))
toxic_model.add(Dense(100, activation='relu'))
toxic_model.add(Dense(50, activation='relu'))

toxic_model.add(Dense(units=2, activation='softmax'))

toxic_model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

toxic_model.summary()

toxic_model.fit(
    X,
    y_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)

toxicP_y = toxic_model.predict_classes(X)
y.sum(), toxicP_y.sum()
```
